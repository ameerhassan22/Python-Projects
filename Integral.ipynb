# Integral as the area under a curve

Although this is a simple example, it demonstrates some important tweaks:

A simple line plot with custom color and line width.

A shaded region created using a Polygon patch.

A text label with mathtext rendering.

figtext calls to label the x- and y-axes.

Use of axis spines to hide the top and right spines.

Custom tick placement and labels.

```python
#call libraries of matplotlib.pyplot , matplotlib.patches-->ploygon  and numpy
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

def func(x):
    return (x - 3) * (x - 5) * (x - 7) + 85    # equation of Integral : X^3 - 15*X^2 + 71*X - 20

a, b = 2, 9                                    # integral limits that a = 2 and b = 9
x = np.linspace(0, 10)                         # random values <= 10
y = func(x)                                    # y = equation of Integral (x - 3) * (x - 5) * (x - 7) = X^3 - 15*X^2 + 71*X - 20

fig, ax = plt.subplots()                       #---> subplots take x-axis , y-axis , colour of line , length of line to draw
ax.plot(x, y, 'g', linewidth=2) 
ax.set_ylim(bottom=1)                          # x value to start  draw equation of Integral

ix = np.linspace(a, b)                    # find value of x from 2 to 9 in equation of Integral (x - 3) * (x - 5) * (x - 7) + 85
iy = func(ix)                             # when x = 2 , y = 20 and when x = 9 , y = 133
verts = [(a, 0), *zip(ix, iy), (b, 0)]   # point (2,0) , (0,9) ,(xi,iy) to draw equation of Integral figure as list


poly = Polygon(verts, facecolor='0.8', edgecolor='0.5')  # polygon  take facecolour , edgecolour , list pf verts
ax.add_patch(poly)                                       # use add_patch

ax.text(0.5 * (a + b), 30, r"$\int_a^b f(x)\mathrm{d}x$",horizontalalignment='center', fontsize=10) 
# text take start point , far of speed under figure about x ,speed under figure , horizontalalignment , fontsize

fig.text(0.9,0.05, '$x$')     # point of x edge of x-axis
fig.text(0.1, 0.9, '$y$')   # point of y edge of y-axis

ax.spines[['top', 'right']].set_visible(False)  # location of x-axis , y-axis at equation of Integral and show in regtanle of x-axis , y-axis
ax.set_xticks([a, b], labels=['$a$', '$b$'])    # location of a , b under of draw equation of Integral
ax.set_yticks([])                               # to not shaw y-axis values
plt.show()                                      # to draw figure
```
